# Time-Series Forecasting with Machine Learning Models


This notebook demonstrates time-series forecasting using ARIMA, SARIMA, and a third model. 
We use the dataset and focus only on data preprocessing and machine learning models without 
extensive data analysis. Each step is explained in detail for better understanding.


```

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

# Load the dataset (replace 'your_dataset.csv' with the actual dataset path)
data_path = '/kaggle/input/your-dataset.csv'
data = pd.read_csv(data_path)

# Display the first few rows of the dataset
data.head()
```


## Data Preprocessing

1. **Handling Missing Values**: Ensure there are no missing values in the dataset.
2. **Formatting the Date Column**: If the dataset contains dates, parse them appropriately.
3. **Setting Up Time-Series Data**: Make sure the data is sorted and indexed by the date.


```

# Handling missing values (if any)
data = data.dropna()

# Convert 'Date' column to datetime (replace 'Date' with the actual column name)
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Plot the time-series data to visualize trends
plt.figure(figsize=(12, 6))
plt.plot(data['Value'])  # Replace 'Value' with the target column name
plt.title('Time-Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```


## ARIMA Model

ARIMA (Auto-Regressive Integrated Moving Average) is a popular statistical method for time-series forecasting.
Here, we'll set up and train an ARIMA model to predict future values.


```

# Define ARIMA model (replace p, d, q with appropriate parameters)
model_arima = ARIMA(data['Value'], order=(5, 1, 0))  # Example parameters
arima_result = model_arima.fit()

# Forecast future values
forecast_arima = arima_result.forecast(steps=10)  # Forecast next 10 steps
print("ARIMA Forecast:", forecast_arima)

# Plot original data and forecast
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Value'], label='Original')
plt.plot(pd.date_range(data.index[-1], periods=10, freq='D'), forecast_arima, label='ARIMA Forecast', color='orange')
plt.legend()
plt.show()
```


## SARIMA Model

SARIMA (Seasonal ARIMA) extends ARIMA by considering seasonality. 
We'll configure a SARIMA model to forecast values considering any observed seasonal patterns.


```

# Define SARIMA model (replace p, d, q, P, D, Q, m with appropriate parameters)
model_sarima = SARIMAX(data['Value'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_result = model_sarima.fit()

# Forecast future values
forecast_sarima = sarima_result.forecast(steps=10)
print("SARIMA Forecast:", forecast_sarima)

# Plot original data and forecast
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Value'], label='Original')
plt.plot(pd.date_range(data.index[-1], periods=10, freq='D'), forecast_sarima, label='SARIMA Forecast', color='green')
plt.legend()
plt.show()
```


## Prophet Model

Prophet is a forecasting tool developed by Facebook. It is robust to missing data and outliers and is ideal for seasonality.


```

from fbprophet import Prophet

# Prepare data for Prophet
prophet_data = data.reset_index()
prophet_data.columns = ['ds', 'y']  # Prophet requires these column names

# Define and fit the model
prophet_model = Prophet()
prophet_model.fit(prophet_data)

# Make future dataframe and forecast
future = prophet_model.make_future_dataframe(periods=10)  # Next 10 days
forecast = prophet_model.predict(future)

# Plot forecast
fig = prophet_model.plot(forecast)
plt.show()
```
